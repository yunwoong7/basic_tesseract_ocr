````python
import pytesseract
import cv2
import matplotlib.pyplot as plt
````

````python
path = 'asset/images/ocr_test.jpg'
````

````python
image = cv2.imread(path)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
````

````python
plt.figure(figsize=(8, 5))

plt.subplot(1, 1, 1)
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.xticks([]), plt.yticks([])

plt.show()
````

```"Failed loading language 'kor'"``` 오류가 발생한다면 'kor' traineddata 가 없어서 발생한 문제입니다.  
[tessdata Download][tessdatalink] 사이트로 이동하여 필요한 언어를 다운로드 후 ```Tesseract-OCR\tessdata``` 경로에 복사하시면 됩니다.

**2개 이상의 언어 사용**이 필요하다면 ```lang='kor+eng'``` 형태로 입력하시면 됩니다.

[tessdatalink]: https://github.com/tesseract-ocr/tessdata "Go tessdata Download"

````python
# use Tesseract to OCR the image
text = pytesseract.image_to_string(rgb_image, lang='kor+eng')
print(text)
````
